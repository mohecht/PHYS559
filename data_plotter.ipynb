```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = #insert filepath here
df = pd.read_csv(filename, header = None, skiprows=[0,1,2,3,4], sep = '    ')
```

```python
x = np.array(df)
rows,columns = np.shape(x)

voltage = x[1:,3]
current = x[1:,2]
time = x[1:,1]

```

```python
fig, axs= plt.subplots(2,sharex = True)
fig.suptitle('DC/SFQ Converter Simulated Responses')
axs[1].set_xlabel('Time (ns)')
axs[0].plot(time*1e9,current*1e6)
axs[0].set_ylabel(r'Input Current ($\mu$A)')
axs[1].plot(time*1e9,voltage*1e3)
axs[1].set_ylabel('Output Voltage (mV)')
```
